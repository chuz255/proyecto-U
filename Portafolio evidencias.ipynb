<ol>
    <li>Verificar el dominio(INT,OBJECT) de cada una de las variables si es necesario se ajusta</li>

<h4 Style = "color:#A0522D"> 3.1 Verificar el dominio de las variables </h4> 

```python
dfAutos.dtypes
```

```python
dfAutos.head(5)
```

```python
dfAutos[['normalized','horsepower','peak','price']] = dfAutos[['normalized','horsepower','peak','price']].astype("int")
dfAutos.dtypes

dfAutos[['bore','stroke']] = dfAutos[['bore','stroke']].astype("float")
dfAutos.dtypes
```

<h4 Style = "color:#A0522D"> 3.2 Estandarizacion de datos </h4> 

```python
dfAutos["City L/100km(calc)"] = 235/dfAutos["city"]
dfAutos["Higway L/100km(calc)"] = 235/dfAutos["higway"]
dfAutos.head(20)
```

<h4 Style = "color:#A0522D"> 3.3 Volver variables comparables </h4> 

```python
dfTemp = dfAutos[["horsepower","peak"]]
dfTemp.head(3)
```

```python
dfAutos["horsepower(norm)"] = dfAutos["horsepower"]/dfAutos["horsepower"].max()
dfAutos["peak(norm)"] = dfAutos["peak"]/dfAutos["peak"].max()
dfAutos.head(5)
```

<h4 Style = "color:#A0522D"> 3.4 Binning </h4> 

```python
dfTemp.head(5)
```

```python
bins = np.linspace(48,262,4)
bins
```

```python
categoria = ["bajo","medio","alto"]
dfAutos["Horsepower(Bins)"] = pd.cut(dfAutos["horsepower"],bins, labels = categoria, include_lowest=True )

dfAutos[["horsepower","Horsepower(Bins)"]].head(5)
```
